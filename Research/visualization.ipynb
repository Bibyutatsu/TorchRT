```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

import pandas as pd
```

```python
torNet_epoch = pd.concat([pd.read_csv('./lightning_logs/run-version_0-tag-epoch.csv'), pd.read_csv('./lightning_logs/run-version_2-tag-epoch.csv')]).assign(Net="torNet")
torNet_trainloss = pd.concat([pd.read_csv('./lightning_logs/run-version_0-tag-train_loss.csv'), pd.read_csv('./lightning_logs/run-version_2-tag-train_loss.csv')]).assign(Net="torNet")[:760]
torNet_valacc = pd.concat([pd.read_csv('./lightning_logs/run-version_0-tag-val_acc.csv'), pd.read_csv('./lightning_logs/run-version_2-tag-val_acc.csv')]).assign(Net="torNet")[:20]
```

```python
fashionCNN_trainloss = pd.read_csv('./lightning_logs/run-version_4-tag-train_loss.csv').assign(Net="FashionCNN")
fashionCNN_valacc = pd.read_csv('./lightning_logs/run-version_4-tag-val_acc.csv').assign(Net="FashionCNN")
```

```python
train_loss = pd.concat([fashionCNN_trainloss, torNet_trainloss])
train_loss.columns = ['Wall time', 'Step', 'Train loss', 'Net']

valacc = pd.concat([fashionCNN_valacc, torNet_valacc])
valacc.columns = ['Wall time', 'Step', 'Val Accuracy', 'Net']
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x="Step", y="Train loss", hue="Net", data=train_loss)
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x="Step", y="Val Accuracy", hue="Net", data=valacc)
```
